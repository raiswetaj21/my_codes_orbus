```python
import fitz  # PyMuPDF
import re

def extract_section_from_pdf(pdf_path, section_title):
    # Open the PDF file
    doc = fitz.open(pdf_path)
    section_content = ""
    is_in_section = False
    
    # Define a regex pattern to match the section title
    title_pattern = re.compile(rf"^{section_title}\b", re.IGNORECASE)
    
    for page_num in range(doc.page_count):
        page = doc[page_num]
        text = page.get_text("text")
        
        # Split the page text into lines
        lines = text.split('\n')
        
        for line in lines:
            # Check if the line matches the section title pattern
            if title_pattern.search(line):
                is_in_section = True
                continue  # Start capturing from the next line
            
            # If we're in the section, capture lines until we reach an empty line
            if is_in_section:
                if line.strip() == "":  # Stop capturing on an empty line
                    is_in_section = False
                    break
                section_content += line + "\n"
        
        # If we've captured the section, we can stop searching
        if section_content:
            break

    doc.close()
    return section_content.strip() if section_content else None

# Example usage
pdf_path = "tenderset.pdf"
section_title = "Scope of Work"
section_text = extract_section_from_pdf(pdf_path, section_title)

if section_text:
    print("Extracted Section Content:-\n", section_text)
else:
    print(f"Section '{section_title}' not found in the document.")
```

```python
import fitz  # PyMuPDF
import re

def extract_full_section_from_pdf(pdf_path, section_title):
    # Open the PDF file
    doc = fitz.open(pdf_path)
    section_content = ""
    is_in_section = False
    
    # Define a regex pattern to match the section title
    title_pattern = re.compile(rf"^{section_title}\b", re.IGNORECASE)
    # Pattern to detect any new title (assuming titles are capitalized or have certain formatting)
    any_title_pattern = re.compile(r"^[A-Z][A-Za-z\s]*:$")

    for page_num in range(doc.page_count):
        page = doc[page_num]
        text = page.get_text("text")
        
        # Split the page text into lines
        lines = text.split('\n')
        
        for line in lines:
            # Check if the line matches the specified section title
            if title_pattern.search(line):
                is_in_section = True  # Start capturing content
                section_content = ""  # Reset content for the last occurrence
                continue
            
            # Stop capturing if another title appears after starting to capture
            if is_in_section and any_title_pattern.search(line):
                is_in_section = False
                break
            
            # Capture lines if we are within the section
            if is_in_section:
                section_content += line + "\n"
                
    doc.close()
    return section_content.strip() if section_content else None

# Example usage
pdf_path = "tenderset.pdf"
section_title = "Overview"
section_text = extract_full_section_from_pdf(pdf_path, section_title)

if section_text:
    print("Extracted Section Content:\n", section_text)
else:
    print(f"Section '{section_title}' not found in the document.")
```

```python
import fitz  # PyMuPDF
import re

def extract_section_content(pdf_path, section_title):
    # Load PDF
    doc = fitz.open(pdf_path)
    content = ""
    start_extracting = False
    
    # Define regex to detect titles (assumes titles are in uppercase or specific formats)
    title_pattern = re.compile(r'^[A-Z ]{5,}$')
    
    for page_num in range(doc.page_count):
        page = doc.load_page(page_num)
        text = page.get_text("text")
        
        # Split page text into lines for processing
        lines = text.splitlines()
        
        for line in lines:
            # Start extracting when section title is found
            if section_title.lower() in line.lower():
                start_extracting = True
            
            # Stop extracting if another title is encountered
            elif start_extracting and title_pattern.match(line.strip()):
                start_extracting = False
                break
            
            # Add line to content if extraction has started
            if start_extracting:
                content += line + "\n"
    
    return content.strip()

# Usage example
pdf_path = "tenderset.pdf"
section_title = "Overview"
extracted_content = extract_section_content(pdf_path, section_title)

# Display or save the extracted content
print(extracted_content)
```
