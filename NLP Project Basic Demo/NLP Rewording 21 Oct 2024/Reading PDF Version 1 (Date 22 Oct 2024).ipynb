```python
import pdfplumber
```

```python
def extract_text_after_title(pdf_path, title):
    extracted_text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text = page.extract_text()
            if text:
                # Split text by lines
                lines = text.split('\n')
                capture = False
                for line in lines:
                    # Check if the line contains the title
                    if title.lower() in line.lower():
                        capture = True
                        continue
                    # If capture is True, start capturing lines below the title
                    if capture:
                        if line.strip() == "":  # Break on encountering an empty line
                            break
                        extracted_text += line + "\n"
    return extracted_text
```

```python
# Usage code chunk on ISWM tender pdf file
pdf_path = "tenderset.pdf"
title = "Overview"
text_below_title = extract_text_after_title(pdf_path, title)

print("Extracted Text:-\n", text_below_title)
```
