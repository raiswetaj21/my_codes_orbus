```python
import numpy as np
import pandas as pd
import random
import faker
from faker import Faker
```

# Projects Table:

```python
# Initilize the package into a variable
fake = Faker()

# Dataframe Record Number Constant
count = 50

def generate_project_data(num_projects):
    # Generate Project Data
    project_data = {
        "ProjectID": [f"PRJ-{i:03d}" for i in range(1, num_projects+1)],
        "ProjectName": [fake.company() for _ in range(num_projects)],
        "ProjectManager": [fake.name() for _ in range(num_projects)],
        "StartDate": [fake.date_between(start_date="-2y", end_date="today") for _ in range(num_projects)],
        "EndDate": [fake.date_between(start_date="today", end_date="+2y") for _ in range(num_projects)],
        "Budget": [round(random.uniform(50000, 1000000), 2) for _ in range(num_projects)],
        "ProjectStatus": [random.choice(['In Progress', 'Completed', 'Not Started']) for _ in range(num_projects)]
    }

    # Create DataFrame and save to CSV
    df_projects = pd.DataFrame(project_data)
#     df_projects.to_csv("projects.csv", index=False)
    
    return df_projects

# Function usage
df_projects = generate_project_data(count)
df_projects
```

# Tasks Table:

```python
def generate_task_data(num_tasks, num_projects):
    # Generate Task Data
    task_data = {
        "TaskID": [f"TSK-{i:04d}" for i in range(1, num_tasks+1)],
        "ProjectID": [f"PRJ-{random.randint(1, num_projects):03d}" for _ in range(num_tasks)],
        "TaskName": [fake.catch_phrase() for _ in range(num_tasks)],
        "AssignedTo": [fake.name() for _ in range(num_tasks)],
        "DueDate": [fake.date_between(start_date="today", end_date="+1y") for _ in range(num_tasks)],
        "TaskStatus": [random.choice(['Pending', 'In Progress', 'Completed']) for _ in range(num_tasks)]
    }

    # Create DataFrame
    df_tasks = pd.DataFrame(task_data)
#     df_tasks.to_csv("tasks.csv", index=False)
    
    return df_tasks

# Function usage
df_tasks = generate_task_data(count, count)
df_tasks
```

# Quotation Table:
